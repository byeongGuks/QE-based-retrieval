```python
import json
from operator import itemgetter

input_file_path = "./dataset/NQ_test_top100_all_QE.json"

benchmark_set = []
with open(input_file_path, 'r', encoding='utf-8') as file :
    data = json.load(file)
    for instance in data :
        candidate_docs = instance['ctxs']
        candidate_docs = sorted(candidate_docs, key=itemgetter('score'), reverse=True)
        
        has_answer = False
        for doc in candidate_docs :
            if doc['has_answer'] :
                has_answer = True
                continue
        if has_answer == False :
            continue
        
        is_well_retrieved = False
        for doc in candidate_docs[:1] :
            if doc['has_answer'] :
                is_well_retrieved = True
                continue
        if is_well_retrieved == True:
            continue
        
        benchmark_set.append(instance)

output_file_path = "./dataset/NQ_test_benchmark_previous_ver.json"  
with open(output_file_path, 'w', encoding='utf-8') as file :
    json.dump(benchmark_set, file)
```
