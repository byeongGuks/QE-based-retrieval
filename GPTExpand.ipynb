```python
#파일들 합치기

import json
import os

#filenames=['NQ_test_benchmark_withgpt_part_1.json','NQ_test_benchmark_withgpt_part_2.json','NQ_test_benchmark_withgpt_part_3.json','NQ_test_benchmark_withgpt_part_4.json']
def aggregate_json_files():
    aggregated_data = []

    # Iterate over all JSON files in the directory
    for filename in ['NQ_test_benchmark_withgpt_part_1.json','NQ_test_benchmark_withgpt_part_2.json','NQ_test_benchmark_withgpt_part_3.json','NQ_test_benchmark_withgpt_part_4(2).json']:
            with open(filename, 'r') as file:
                json_data = json.load(file)
                # Add the data from the current file to the aggregated data
                aggregated_data.extend(json_data)

    return aggregated_data

# Call the function to aggregate the JSON files
aggregated_data = aggregate_json_files()

# Write the aggregated data to a new JSON file

with open('NQ_test_benchmark_withgpt_combined.json', 'w') as output_file:
    json.dump(aggregated_data, output_file)
```

```python
#중복된 query 제거

from collections import defaultdict

def find_duplicate_sentences(sentences):
    duplicates = defaultdict(list)
    sentence_index = {}

    # Iterate over the sentences
    for index, sentence in enumerate(sentences):
        # Remove leading/trailing whitespace and convert to lowercase for comparison
        sentence = sentence.strip().lower()

        # Check if the sentence has already been encountered
        if sentence in sentence_index:
            # If yes, it's a duplicate
            duplicates[sentence].append(index)
        else:
            # If no, store the index of the first occurrence
            sentence_index[sentence] = index

    return duplicates

# Example usage
#text = "This is a sample sentence. This sentence is similar. Another sentence is similar to the previous one. This is a sample sentence."
#sentences = text.split(". ")
sentences = ['a' for _ in range(319)]
for i in range(0,319):
  sentences[i]=dataa[i]["question"]
  
duplicate_sentences = find_duplicate_sentences(sentences)

# Print the duplicate sentences and their indices
for sentence, indices in duplicate_sentences.items():
    print(f"Sentence '{sentence}' is duplicated at indices: {', '.join(map(str, indices))}")
```
